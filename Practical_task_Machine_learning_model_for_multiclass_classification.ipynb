```python

# 0.  One‑off installs  &  global reproducibility
!pip -q install optuna xgboost numbers-parser kaleido

import warnings, os, random, pathlib, sys, gc
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
sns.set_style("whitegrid")

SEED = 42
random.seed(SEED);  np.random.seed(SEED);  os.environ["PYTHONHASHSEED"]=str(SEED)
warnings.filterwarnings("ignore")

import importlib.metadata as md
print("Versions:", {p: md.version(p) for p in
      ["numpy","pandas","scikit-learn","optuna","xgboost"]})

# 1.  Data upload  (CSV, XLSX, or Apple Numbers)
from google.colab import files
up_name = next(iter(files.upload()))
ext     = pathlib.Path(up_name).suffix.lower()

if ext == ".numbers":
    from numbers_parser import Document
    doc   = Document(up_name)
    sheet = doc.sheets[0]; table = sheet.tables[0]
    rows  = list(table.rows(values_only=True))
    df    = pd.DataFrame(rows[1:], columns=rows[0]).rename(str.strip, axis=1)
elif ext in (".xlsx", ".xls"):
    df = pd.read_excel(up_name).rename(str.strip, axis=1)
else:                                  # assume CSV
    df = pd.read_csv(up_name).rename(str.strip, axis=1)

assert {'MMSE','GM_score','Diagnosis'}.issubset(df.columns)
df.drop(columns=[c for c in df if c.upper()=='RID'], inplace=True, errors='ignore')

classes = sorted(df['Diagnosis'].unique())        # ['CN','Dementia','MCI']
y_raw   = pd.Categorical(df['Diagnosis'], categories=classes).codes
X_raw   = df[['MMSE','GM_score']].copy()          # keep DataFrame

print(f"data shape {df.shape}   classes {dict(enumerate(classes))}")

plt.figure(figsize=(4,3))
sns.countplot(x=df['Diagnosis'], order=classes, palette="Blues_r")
plt.title("Class distribution"); plt.ylabel("# patients"); plt.tight_layout()
plt.show()

# 2.  Train / val / test  (60 / 20 / 20)
from sklearn.model_selection import train_test_split
X_tr, X_te, y_tr, y_te = train_test_split(
        X_raw, y_raw, test_size=.20, stratify=y_raw, random_state=SEED)
X_tr, X_va, y_tr, y_va = train_test_split(
        X_tr,  y_tr,  test_size=.25, stratify=y_tr,  random_state=SEED)

# 3.  Helper funcs  (conf‑mat, ROC, metrics)
from sklearn.metrics import (confusion_matrix, balanced_accuracy_score,
                             accuracy_score, f1_score, roc_curve, auc)
import plotly.io as pio, optuna.visualization as oviz

def show_cm(y_true, y_pred, title):
    cm  = confusion_matrix(y_true, y_pred)
    cmn = cm / cm.sum(1, keepdims=True)
    fig, ax = plt.subplots(1,2,figsize=(9,4))
    sns.heatmap(cm ,annot=True,fmt='d',cmap='Blues',
                xticklabels=classes,yticklabels=classes,ax=ax[0])
    sns.heatmap(cmn,annot=True,fmt='.2f',cmap='Blues',
                xticklabels=classes,yticklabels=classes,ax=ax[1])
    ax[0].set_title(title+" counts");  ax[1].set_title(title+" (%)")
    for a in ax: a.set_xlabel("pred"); ax[0].set_ylabel("true")
    plt.tight_layout(); plt.show()

def roc_ovr(model, X, y, title):
    if not hasattr(model,"predict_proba"): return
    P   = model.predict_proba(X)
    plt.figure(figsize=(6,5)); plt.plot([0,1],[0,1],'k--',lw=1)
    aucs=[]
    for i,c in enumerate(classes):
        fpr,tpr,_ = roc_curve((y==i).astype(int), P[:,i])
        A = auc(fpr,tpr); aucs.append(A)
        plt.plot(fpr,tpr,lw=2,label=f"{c} (AUC={A:.2f})")
    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title(title+" – ROC (OvR)")
    plt.legend(); plt.grid(alpha=.3); plt.tight_layout(); plt.show()
    print(f"Macro‑avg AUC = {np.mean(aucs):.3f}")

def evaluate(tag, model, X, y, *, store=None, key=None):
    """Print + plots  + store metrics if dict provided."""
    pr  = model.predict(X)
    ba  = balanced_accuracy_score(y, pr)
    acc = accuracy_score(y, pr)
    f1  = f1_score(y, pr, average="macro")
    print(f"{tag:18} acc {acc:.3f}  BA {ba:.3f}  F1‑m {f1:.3f}")
    show_cm(y, pr, tag);  roc_ovr(model, X, y, tag)
    if store is not None:
        store[key if key else tag] = dict(acc=acc, bal_acc=ba, f1=f1)

# 4.  Pre‑processing & feature engineering
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, FunctionTransformer
from sklearn.compose       import ColumnTransformer
from sklearn.pipeline      import Pipeline

def add_ratio(X):
    X = X.copy()
    X['MMSE_over_GM'] = X['MMSE'] / (X['GM_score'] + 1e-6)
    return X

preprocess = Pipeline([
    ('ratio',  FunctionTransformer(add_ratio)),
    ('selector', ColumnTransformer([
            ('orig',  'passthrough', ['MMSE','GM_score']),
            ('poly',  PolynomialFeatures(2, include_bias=False), ['MMSE','GM_score']),
            ('ratio', 'passthrough', ['MMSE_over_GM'])
        ], remainder='drop')),
    ('sc', StandardScaler())
])

# 5.  Models & Optuna tuning
from sklearn.linear_model  import LogisticRegression
from sklearn.svm           import SVC
from sklearn.ensemble      import RandomForestClassifier, GradientBoostingClassifier
from xgboost               import XGBClassifier
import optuna
from sklearn.model_selection import StratifiedKFold, cross_val_score

def pipe(clf): return Pipeline([('prep', preprocess), ('clf', clf)])

MODELS = {
    "LogReg": pipe(LogisticRegression(max_iter=3000,class_weight='balanced',
                                      random_state=SEED)),
    "SVC"   : pipe(SVC(kernel='rbf',probability=True,class_weight='balanced',
                       random_state=SEED)),
    "RF"    : pipe(RandomForestClassifier(n_estimators=500,class_weight='balanced',
                       random_state=SEED)),
    "GB"    : pipe(GradientBoostingClassifier(random_state=SEED)),
    "XGB"   : pipe(XGBClassifier(n_estimators=400,learning_rate=0.05,max_depth=4,
                       subsample=.8,objective='multi:softprob',num_class=3,
                       eval_metric='mlogloss',random_state=SEED))
}

def tune_svc(trials=50):
    cv = StratifiedKFold(3, shuffle=True, random_state=SEED)
    def objective(trial):
        m = pipe(SVC(probability=True,class_weight='balanced',
                     C = trial.suggest_float("C",1e-2,1e2,log=True),
                     gamma = trial.suggest_float("gamma",1e-3,10,log=True),
                     random_state=SEED))
        return cross_val_score(m, X_tr, y_tr, cv=cv,
                               scoring="balanced_accuracy").mean()
    study = optuna.create_study(direction="maximize",
                                sampler=optuna.samplers.TPESampler(seed=SEED))
    study.optimize(objective, n_trials=trials, show_progress_bar=False)
    print("SVC best:", study.best_params)
    oviz.plot_optimization_history(study).show()
    oviz.plot_param_importances(study).show()
    return pipe(SVC(probability=True,class_weight='balanced',
                    random_state=SEED, **study.best_params))

def tune_xgb(trials=50):
    cv = StratifiedKFold(3, shuffle=True, random_state=SEED)
    def objective(trial):
        p = dict(n_estimators = trial.suggest_int("n_est",200,500),
                 max_depth    = trial.suggest_int("depth",3,6),
                 learning_rate= trial.suggest_float("lr",0.01,0.3,log=True),
                 subsample    = trial.suggest_float("ss",0.6,1.0),
                 colsample_bytree= trial.suggest_float("cs",0.6,1.0))
        m = pipe(XGBClassifier(**p,objective='multi:softprob',num_class=3,
                               eval_metric='mlogloss',random_state=SEED))
        return cross_val_score(m, X_tr, y_tr, cv=cv,
                               scoring="balanced_accuracy").mean()
    study = optuna.create_study(direction="maximize",
                                sampler=optuna.samplers.TPESampler(seed=SEED))
    study.optimize(objective, n_trials=trials, show_progress_bar=False)
    print("XGB best:", study.best_params)
    oviz.plot_optimization_history(study).show()
    oviz.plot_param_importances(study).show()
    p = study.best_params
    p.update(dict(n_estimators=p.pop('n_est'), max_depth=p.pop('depth'),
                  learning_rate=p.pop('lr'),  subsample=p.pop('ss'),
                  colsample_bytree=p.pop('cs')))
    return pipe(XGBClassifier(**p,objective='multi:softprob',num_class=3,
                              eval_metric='mlogloss',random_state=SEED))

MODELS["SVC_tuned"] = tune_svc()
MODELS["XGB_tuned"] = tune_xgb()

# Platt scaling
from sklearn.calibration import CalibratedClassifierCV
base_svc = MODELS["SVC_tuned"].named_steps['clf']
MODELS["SVC_cal"] = pipe(CalibratedClassifierCV(estimator=base_svc, method='sigmoid', cv=3))

# 6.  Fit each model – FULL visual output  (val & test)
metrics_val, metrics_test = {}, {}
for name, mdl in MODELS.items():
    print(f"\n{'═'*30}\n{ name } – VALIDATION")
    mdl.fit(X_tr, y_tr)
    evaluate(f"{name} (val)",  mdl, X_va, y_va,
             store=metrics_val, key=name)
    print(f"{'-'*30}\n{ name } – TEST")
    evaluate(f"{name} (test)", mdl, X_te, y_te,
             store=metrics_test, key=name)

# 7.  Soft‑vote ensemble  +  τ* on MCI
top3     = sorted(metrics_val.items(), key=lambda kv:kv[1]['bal_acc'],
                  reverse=True)[:3]
names_top= [n for n,_ in top3]
weights  = [metrics_val[n]['bal_acc'] for n in names_top]

from sklearn.ensemble import VotingClassifier
ensemble = VotingClassifier([(n, MODELS[n]) for n in names_top],
                            voting='soft', weights=weights).fit(X_tr, y_tr)

idx_MCI = classes.index("MCI")
grid    = np.arange(0.15,0.90,0.02)
best_ba, best_thr = 0, 0.5
P_va = ensemble.predict_proba(X_va)
for t in grid:
    pr = np.where(P_va[:, idx_MCI] >= t, idx_MCI, P_va.argmax(1))
    ba = balanced_accuracy_score(y_va, pr)
    if ba > best_ba: best_ba, best_thr = ba, t
print(f"\nBest τ on validation: {best_thr:.2f}  (BA={best_ba:.3f})")

plt.figure(figsize=(5,3))
plt.plot(grid, [balanced_accuracy_score(
        y_va, np.where(P_va[:, idx_MCI]>=t, idx_MCI, P_va.argmax(1))) for t in grid])
plt.axvline(best_thr, color='r', ls='--'); plt.xlabel("MCI prob threshold")
plt.ylabel("Balanced accuracy"); plt.tight_layout(); plt.show()

print("\nEnsemble – TEST (default arg‑max):")
evaluate("Ensemble_default (test)", ensemble, X_te, y_te)

print("\nEnsemble – TEST (MCI tuned):")
te_pred = np.where(ensemble.predict_proba(X_te)[:, idx_MCI]>=best_thr,
                   idx_MCI, ensemble.predict(X_te))
evaluate("Ensemble_MCI‑tuned (test)", ensemble, X_te, y_te*0+te_pred)

# 8.  Heat‑maps (inline) – quick summary tables
def heat(df, title):
    plt.figure(figsize=(6,4))
    sns.heatmap(df.sort_values('bal_acc',ascending=False),
                annot=True, fmt='.3f', cmap='YlGnBu')
    plt.title(title); plt.tight_layout(); plt.show()

heat(pd.DataFrame(metrics_val ).T[['bal_acc','acc','f1']],
     "VALIDATION – model comparison")
heat(pd.DataFrame(metrics_test).T[['bal_acc','acc','f1']],
     "TEST – model comparison")
```
